## 1. Variables = Mise en Place

**Kitchen:** Before service starts, a chef measures and labels every ingredient in prep bowls — butter here, flour there, eggs over here. Each bowl has a name and holds a value.

**Code:** A variable is a labeled container. You assign a value to a name, then use that name whenever you need the value.

````python
# === MISE EN PLACE — set out your ingredients ===

# Assignment: label a bowl, put a value in it
butter = 2       # cups
flour = 3        # cups
sugar = 0.5      # cups
oven_temp = 350  # degrees F

print("Prep bowls ready:")
print(f"  butter  = {butter} cups")
print(f"  flour   = {flour} cups")
print(f"  sugar   = {sugar} cups")
print(f"  oven    = {oven_temp}°F")

# Usage: reference by name — the value follows
dough_volume = butter + flour + sugar
print(f"\nTotal dough volume: {dough_volume} cups")

# Change the value once at the top — it updates everywhere
butter = 4
dough_volume = butter + flour + sugar
print(f"Doubled butter → new dough volume: {dough_volume} cups")
````

## 2. Functions = Recipes

**Kitchen:** A recipe is a reusable set of steps. Given ingredients (inputs), it produces a dish (output). Same recipe, different quantities each time.

**Code:** A function takes parameters, runs a body of code, and returns a result. Arguments are the specific values you pass in.

````python
# === THE RECIPE — define it once, use it with different ingredients ===

def make_cookies(flour, butter, sugar, temp):
    """A cookie recipe: mix, shape, bake, return the batch."""
    
    # Step 1: mix dry ingredients (uses flour + sugar)
    dry_mix = flour + sugar
    print(f"  Mixed dry ingredients: {dry_mix} cups total")
    
    # Step 2: cream butter into the mix (uses butter)
    dough = dry_mix + butter
    print(f"  Creamed in {butter} cups butter → {dough} cups dough")
    
    # Step 3: bake at the given temp (uses temp)
    print(f"  Baking at {temp}°F...")
    
    # Step 4: return the finished product
    batch_size = int(dough * 12)  # ~12 cookies per cup of dough
    return batch_size


# --- Call the recipe with specific arguments ---
print("Batch 1 (small):")
cookies = make_cookies(flour=2, butter=1, sugar=0.5, temp=350)
print(f"  → {cookies} cookies\n")

# Same recipe, different arguments → different output
print("Batch 2 (large):")
cookies = make_cookies(flour=4, butter=2, sugar=1, temp=375)
print(f"  → {cookies} cookies")
````

## 3. Loops = Batch Prep

**Kitchen:** Peeling 50 potatoes is the same action repeated for each item. You don't write the instructions 50 times — you just say "do this for each potato in the bin."

**Code:** A `for` loop runs the same block of code once for every item in a collection.

````python
# === BATCH PREP — same action, every item in the bin ===

vegetables = ["onion", "carrot", "celery", "garlic", "leek"]

def chop(item):
    return f"{item} chopped"

def season(item):
    return f"{item} seasoned"

# Write the action once — the loop handles the rest
print("Prep station running:\n")

prepped = []
for item in vegetables:
    result = chop(item)
    result = season(item)
    prepped.append(result)
    print(f"  ✓ {item}: chopped and seasoned")

print(f"\nTotal items prepped: {len(prepped)}")
print("Add more to the list — the loop handles it. No extra code needed.")

# --- Bonus: a while loop (keep cooking until done) ---
print("\n--- While loop: reducing sauce ---")
sauce_cups = 4.0
target = 2.0

while sauce_cups > target:
    sauce_cups -= 0.5
    print(f"  Simmering... {sauce_cups} cups remaining")

print(f"  Done! Sauce reduced to {sauce_cups} cups.")
````

## 4. Conditionals = Line Judgment Calls

**Kitchen:** *"If the fish is at temp, plate it. Otherwise, keep it on the heat."* The chef checks a condition and branches to the right action.

**Code:** `if / elif / else` checks conditions top-to-bottom and runs the first branch that's `True`.

````python
# === LINE JUDGMENT CALLS — check a condition, take one branch ===

def check_fish(fish_temp):
    """Decide what to do with the fish based on its temperature."""
    
    print(f"  Fish temp: {fish_temp}°F")
    
    if fish_temp >= 145:
        action = "Plate and serve — it's perfect!"
    elif fish_temp >= 120:
        action = "Almost there — rest it one more minute."
    else:
        action = "Keep cooking — not ready yet."
    
    print(f"  Decision: {action}")
    return action


# Try different temperatures — only ONE branch runs each time
print("Scenario 1:")
check_fish(150)    # hits the first branch

print("\nScenario 2:")
check_fish(130)    # hits the second branch (elif)

print("\nScenario 3:")
check_fish(95)     # hits the else
````

## 5. Classes & Objects = Station Blueprint

**Kitchen:** A class is the *blueprint* for a station (e.g., the sauté station): it defines what tools it has and what it can do. Each actual station you set up is an *instance* — same design, operating independently.

**Code:** A class defines attributes (data) and methods (actions). Each object created from it has its own independent state.

````python
# === STATION BLUEPRINT — one class, many independent instances ===

class SauteStation:
    """Blueprint for a sauté station in the kitchen."""
    
    def __init__(self, chef):
        self.chef = chef      # who's running this station
        self.temp = 0         # pan starts cold
        self.current_dish = None
    
    def heat_pan(self, degrees):
        self.temp = degrees
        print(f"  {self.chef}'s pan heated to {self.temp}°F")
    
    def cook(self, dish):
        self.current_dish = dish
        print(f"  {self.chef} is cooking {dish} at {self.temp}°F")
    
    def status(self):
        print(f"  Station [{self.chef}]: temp={self.temp}°F, dish={self.current_dish}")


# --- Create two independent stations from the same blueprint ---
station_a = SauteStation("Marco")
station_b = SauteStation("Sofia")

# Each station operates independently
print("Setting up stations:")
station_a.heat_pan(450)
station_a.cook("duck breast")

station_b.heat_pan(380)
station_b.cook("wilted spinach")

# Prove they're independent — changing one doesn't affect the other
print("\nFinal status:")
station_a.status()
station_b.status()
print("\nSame blueprint, totally independent kitchens.")
````

## 6. Scope = Your Station vs Theirs

**Kitchen:** The grill cook doesn't reach over and use the pastry chef's sugar. Each station's ingredients stay at their own station.

**Code:** Variables defined inside a function are *local* — they only exist within that function. Trying to access them from outside causes a `NameError`.

````python
# === STATION BOUNDARIES — local variables stay local ===

# Global variable — visible everywhere (the shared pantry)
kitchen_name = "La Maison"


def grill_protein():
    marinade = "herb oil"       # local to this function
    grill_temp = 550            # local to this function
    print(f"  Grill station at {kitchen_name}")
    print(f"  Using marinade: {marinade}, temp: {grill_temp}")

def make_dessert():
    pastry_cream = "2 cups"     # local to this function
    oven_temp = 325             # local to this function
    print(f"  Pastry station at {kitchen_name}")
    print(f"  Using pastry_cream: {pastry_cream}, oven: {oven_temp}")


# Both functions can see the global kitchen_name
print("Both stations access the shared pantry (global):")
grill_protein()
make_dessert()

# But the grill cook can't reach the pastry chef's pastry_cream
print("\nTrying to access 'pastry_cream' from outside make_dessert():")
try:
    print(pastry_cream)
except NameError as e:
    print(f"  NameError: {e}")
    print("  → 'pastry_cream' lives inside make_dessert() only!")
````

## 8. Debugging = Tasting as You Go

**Kitchen:** A good chef tastes at every stage — after reducing, after seasoning, after emulsifying. If the dish goes wrong at the end, they can trace it back to the exact step.

**Code:** Add `print()` statements to inspect values at each step. Find *exactly* where the value went wrong.

````python
# === TASTING AS YOU GO — print() at each step to find the bug ===

def make_sauce(stock_cups, salt_tsp):
    """A sauce recipe with a sneaky bug. Can you find it?"""
    
    # Step 1: reduce the stock
    reduced = stock_cups / 2
    print(f"  [DEBUG] after reduce: {reduced} cups")   # taste #1
    
    # Step 2: add salt (BUG: accidentally multiplying instead of adding)
    seasoned_salt = reduced * salt_tsp   # should be: reduced + (salt_tsp * 0.1)
    print(f"  [DEBUG] after salt: {seasoned_salt}")     # taste #2 — bug shows here!
    
    # Step 3: emulsify
    finished = seasoned_salt + 0.5  # add butter
    print(f"  [DEBUG] finished sauce: {finished}")
    
    return finished


print("Running the buggy recipe:")
result = make_sauce(stock_cups=4, salt_tsp=3)
print(f"\nFinal result: {result}")
print("Expected ~2.8 cups, got something way off.")
print("The [DEBUG] prints show the problem happened at step 2!")

# --- Now fix it ---
print("\n--- Fixed version ---")

def make_sauce_fixed(stock_cups, salt_tsp):
    reduced = stock_cups / 2
    seasoned_salt = reduced + (salt_tsp * 0.1)  # fixed: add, not multiply
    finished = seasoned_salt + 0.5
    return finished

result = make_sauce_fixed(stock_cups=4, salt_tsp=3)
print(f"Fixed result: {result} cups — that's right!")
````

## 9. Error Handling = The Plan B

**Kitchen:** What if you're out of cream? A prepared kitchen has a contingency — switch to butter.

**Code:** `try/except` attempts the ideal path. If an error occurs, `except` catches it and runs the fallback — the program keeps running instead of crashing.

````python
# === THE PLAN B — try the ideal path, except catches the problem ===

# Simulated pantry — cream is missing!
pantry = {
    "butter": "1 cup",
    "stock": "4 cups",
    "shallots": "3 each",
    # "cream" is NOT in the pantry
}


def make_beurre_blanc():
    """Try cream first, fall back to butter if unavailable."""
    
    try:
        cream = pantry["cream"]       # this will raise KeyError
        print(f"  Using cream: {cream}")
        sauce_base = cream
        
    except KeyError:
        print("  Out of cream! Falling back to butter...")
        butter = pantry["butter"]
        print(f"  Using butter instead: {butter}")
        sauce_base = butter
    
    print(f"  Sauce made with: {sauce_base}")
    return sauce_base


# --- Path B runs (cream missing, butter fallback) ---
print("Making beurre blanc:")
make_beurre_blanc()

# --- Now add cream and try again (Path A) ---
print("\nRestocking cream...")
pantry["cream"] = "1 cup heavy"

print("Making beurre blanc again:")
make_beurre_blanc()

# --- What happens WITHOUT try/except ---
print("\n--- Without error handling: ---")
del pantry["cream"]  # remove cream again
try:
    # This simulates code WITHOUT try/except
    sauce = pantry["cream"]  # CRASH
except KeyError as e:
    print(f"  KeyError: {e} — program would crash here!")
    print("  That's why we always have a Plan B.")
````

## 10. Comments = Recipe Notes in the Margin

**Kitchen:** *"Add salt gradually — this stock runs salty."* Written in the margin, it doesn't change what the recipe does, but tells the next cook what you were thinking.

**Code:** Comments are ignored by Python but essential for humans. Three styles: docstrings, inline comments, and warning notes.

````python
# === RECIPE NOTES IN THE MARGIN ===


def reduce_stock(stock_cups, target_cups):
    """Reduce stock by simmering until it reaches target volume.
    
    This is a DOCSTRING — it explains the function's purpose.
    Shows up when someone calls help(reduce_stock).
    
    Args:
        stock_cups: starting volume in cups
        target_cups: desired final volume
    
    Returns:
        Final volume of reduced stock.
    """
    current = stock_cups  # measure starting volume (INLINE COMMENT)
    
    while current > target_cups:
        # WARNING: stir constantly — stock scorches below medium heat
        current -= 0.5    # each simmer cycle reduces by 0.5 cups
    
    return current  # reduced and ready to use


# --- Run it ---
result = reduce_stock(4, 2)
print(f"Reduced stock: {result} cups")

# --- Proof that the docstring is accessible ---
print("\n--- The docstring is built-in documentation: ---")
help(reduce_stock)
````

## 11. APIs = The Pass-Through Window

**Kitchen:** The dining room places an order through the pass window. They don't know *how* the kitchen made the dish — they just send a request and get a plate back.

**Code:** An API defines a *contract*: send the right request format, get back a predictable response. The caller never sees the internals.

````python
# === THE PASS-THROUGH WINDOW — request in, response out ===

# The "kitchen" (API backend) — dining room never sees this code
class KitchenAPI:
    """Simulates a restaurant API. The dining room interacts
    through defined endpoints only."""
    
    _menu = {
        "salmon": {"price": 28, "prep_time_min": 15, "available": True},
        "steak":  {"price": 42, "prep_time_min": 20, "available": True},
        "risotto":{"price": 22, "prep_time_min": 25, "available": False},
    }
    
    @staticmethod
    def get_menu_item(dish_name):
        """GET /menu/{dish} — returns dish info or an error."""
        if dish_name in KitchenAPI._menu:
            return {"status": 200, "data": KitchenAPI._menu[dish_name]}
        return {"status": 404, "error": f"'{dish_name}' not on menu"}
    
    @staticmethod
    def place_order(dish_name, table_number):
        """POST /orders — places an order if available."""
        item = KitchenAPI._menu.get(dish_name)
        if not item:
            return {"status": 404, "error": "dish not found"}
        if not item["available"]:
            return {"status": 400, "error": f"'{dish_name}' is 86'd"}
        return {
            "status": 201,
            "order": {"dish": dish_name, "table": table_number,
                      "eta_min": item["prep_time_min"]}
        }


# --- The "dining room" (client) — only uses the public API ---
print("Dining room checks the menu:")
response = KitchenAPI.get_menu_item("salmon")
print(f"  {response}")

print("\nDining room places an order:")
response = KitchenAPI.place_order("salmon", table_number=7)
print(f"  {response}")

print("\nDining room tries to order the 86'd risotto:")
response = KitchenAPI.place_order("risotto", table_number=3)
print(f"  {response}")

print("\nThe dining room never saw _menu directly — only the window.")
````

## 12. Testing = The Soft Opening

**Kitchen:** Before opening night, you run a soft opening — friends and family find the problems in a low-stakes environment.

**Code:** `assert` statements and test functions verify that each recipe works correctly *before* you ship it.

````python
# === THE SOFT OPENING — test each recipe before real service ===

# --- The recipes to test ---
def calculate_food_cost(price_per_unit, quantity):
    return round(price_per_unit * quantity, 2)

def calculate_tip(bill_total, tip_percent=20):
    return round(bill_total * (tip_percent / 100), 2)

def split_bill(total, num_people):
    if num_people <= 0:
        raise ValueError("Need at least 1 person to split the bill")
    return round(total / num_people, 2)


# --- Unit tests: test each recipe in isolation ---
def test_food_cost():
    assert calculate_food_cost(12.50, 4) == 50.0, "Basic cost failed"
    assert calculate_food_cost(0, 10) == 0, "Zero price failed"
    print("  ✓ test_food_cost passed")

def test_tip():
    assert calculate_tip(100) == 20.0, "Default 20% failed"
    assert calculate_tip(100, tip_percent=15) == 15.0, "15% failed"
    assert calculate_tip(0) == 0, "Zero bill failed"
    print("  ✓ test_tip passed")

def test_split_bill():
    assert split_bill(100, 4) == 25.0, "Even split failed"
    assert split_bill(100, 3) == 33.33, "Uneven split failed"
    # Test the error case
    try:
        split_bill(100, 0)
        assert False, "Should have raised ValueError"
    except ValueError:
        pass  # this is expected
    print("  ✓ test_split_bill passed")


# --- Integration test: full table order flow ---
def test_full_table_order():
    cost = calculate_food_cost(28, 4)       # 4 salmon dishes
    tip = calculate_tip(cost)               # default 20%
    total = cost + tip
    per_person = split_bill(total, 4)
    assert per_person == 33.6, f"Expected 33.6, got {per_person}"
    print("  ✓ test_full_table_order passed")


# --- Run the soft opening ---
print("Running soft opening tests:\n")
print("Unit tests (one recipe at a time):")
test_food_cost()
test_tip()
test_split_bill()

print("\nIntegration test (full table order):")
test_full_table_order()

print("\nAll tests passed — confident to open for real service!")
````

## 13. Refactoring = Redesigning the Prep Workflow

**Kitchen:** Same dish on the plate — but the path to get there is cleaner, faster, and easier to maintain.

**Code:** Refactoring improves code structure without changing what it does.

````python
# === REDESIGNING THE PREP WORKFLOW ===

# --- BEFORE: repetitive, hard to maintain ---
print("BEFORE refactoring (chaotic prep):")

onion_chopped = True
onion_seasoned = True
carrot_chopped = True
carrot_seasoned = True
celery_chopped = True
celery_seasoned = True
garlic_chopped = True
garlic_seasoned = True

total_before = sum([onion_chopped, onion_seasoned, carrot_chopped,
                    carrot_seasoned, celery_chopped, celery_seasoned,
                    garlic_chopped, garlic_seasoned])
print(f"  {total_before} manual steps, 8 variables to track")
print("  Adding a new vegetable means 2 more variables + 2 more lines\n")


# --- AFTER: clean, extensible, DRY ---
print("AFTER refactoring (clean prep system):")

def prep_vegetable(veg):
    """Chop and season one vegetable."""
    return {"name": veg, "chopped": True, "seasoned": True}

vegetables = ["onion", "carrot", "celery", "garlic"]
prepped = [prep_vegetable(v) for v in vegetables]

for item in prepped:
    print(f"  ✓ {item['name']}: chopped={item['chopped']}, seasoned={item['seasoned']}")

# Adding a new veg is trivial — just extend the list
vegetables.append("leek")
vegetables.append("fennel")
prepped = [prep_vegetable(v) for v in vegetables]
print(f"\n  Added 2 more vegetables — same code, now handles {len(prepped)} items.")
print("  The dish on the plate is identical — the kitchen just runs better.")
````

## 14. Good Coding Practices

A well-run kitchen has standards. So does well-written code.

### 14a. DRY — Don't Repeat Yourself

````python
# === DRY: Don't Repeat Yourself ===

# BAD — same logic copy-pasted
print("BAD (WET — Write Everything Twice):")
salmon_cost = 28 * 1.08   # price * tax
steak_cost = 42 * 1.08    # same tax calc repeated
pasta_cost = 18 * 1.08    # and again...
print(f"  salmon: ${salmon_cost:.2f}, steak: ${steak_cost:.2f}, pasta: ${pasta_cost:.2f}")
print("  Tax rate changes? Fix it in 3 places.\n")

# GOOD — extract the shared logic into a function
print("GOOD (DRY):")

TAX_RATE = 1.08  # constant — change once, updates everywhere

def with_tax(price):
    return round(price * TAX_RATE, 2)

dishes = {"salmon": 28, "steak": 42, "pasta": 18}
for dish, price in dishes.items():
    print(f"  {dish}: ${with_tax(price):.2f}")

print("  Tax rate changes? Fix it in ONE place.")
````

### 14b. Naming Conventions — Labels Should Say What's in the Bowl

````python
# === NAMING: clear names read like a recipe card ===

# BAD — cryptic names
print("BAD naming:")
x = 350
y = 25
z = x * y
print(f"  x={x}, y={y}, z={z}")
print("  What do x, y, z mean? No one knows.\n")

# GOOD — descriptive names
print("GOOD naming:")
oven_temp_f = 350
cook_time_min = 25
total_heat_units = oven_temp_f * cook_time_min
print(f"  oven_temp_f={oven_temp_f}, cook_time_min={cook_time_min}")
print(f"  total_heat_units={total_heat_units}")
print("  Reads like plain English.\n")

# --- Python naming conventions ---
print("Python naming conventions:")

# snake_case for variables and functions
prep_time_minutes = 15
def calculate_food_cost(price, quantity):
    return price * quantity

# PascalCase for classes
class KitchenStation:
    pass

# UPPER_CASE for constants
MAX_OVEN_TEMP = 500
DEFAULT_SALT_TSP = 1

print(f"  variable:  prep_time_minutes = {prep_time_minutes}")
print(f"  function:  calculate_food_cost(28, 4) = {calculate_food_cost(28, 4)}")
print(f"  class:     KitchenStation")
print(f"  constant:  MAX_OVEN_TEMP = {MAX_OVEN_TEMP}")
````

### 14c. Code Layout — A Tidy Kitchen Is a Fast Kitchen

````python
# === LAYOUT: space, structure, one thought per line ===


# BAD — dense, no breathing room
# def make_stock(bones,water,veg,hours):pot=fill(water);add(bones);add(veg);simmer(pot,hours);return strain(pot)


# GOOD — clean layout, easy to scan
def make_stock(bones, water, veg, hours):
    """Simmer bones and vegetables to produce a rich stock."""
    
    # Fill the pot
    pot_contents = [water]
    
    # Add solids
    pot_contents.append(bones)
    pot_contents.append(veg)
    
    # Simmer
    print(f"  Simmering {len(pot_contents)} ingredients for {hours} hours...")
    
    # Strain and return
    stock = f"rich stock from {bones} and {veg}"
    return stock


result = make_stock("chicken bones", "8 cups water", "mirepoix", 4)
print(f"  Result: {result}")

print("\nLayout rules:")
print("  • 4-space indentation (Python standard)")
print("  • Spaces after commas: f(a, b, c) not f(a,b,c)")
print("  • Blank lines between logical sections")
print("  • One action per line — easy to read, easy to debug")
````

### 14d. Single Responsibility — One Job Per Function

````python
# === SINGLE RESPONSIBILITY: each function does ONE thing well ===

# BAD — one monster function doing everything
# def do_everything(order):
#     ... 200 lines of taking order, cooking, plating, charging ...


# GOOD — small focused functions, each with one job
def take_order(table, dish):
    """Records what the table ordered."""
    order = {"table": table, "dish": dish, "status": "received"}
    print(f"  Order received: table {table} wants {dish}")
    return order


def cook_dish(order):
    """Prepares the dish."""
    order["status"] = "cooking"
    print(f"  Cooking {order['dish']}...")
    order["status"] = "ready"
    return order


def plate_dish(order):
    """Plates the finished dish."""
    order["status"] = "plated"
    print(f"  Plating {order['dish']} for table {order['table']}")
    return order


def charge_table(order, price):
    """Processes the payment."""
    order["total"] = price
    order["status"] = "paid"
    print(f"  Table {order['table']} charged ${price}")
    return order


# --- Full service flow: each step is clear and testable ---
print("Full service flow:\n")
order = take_order(table=7, dish="salmon")
order = cook_dish(order)
order = plate_dish(order)
order = charge_table(order, price=28)

print(f"\nFinal order: {order}")
print("\nEach function is small, testable, and does exactly one thing.")
````

```bash
git checkout -b feature/new-risotto
# ... make changes safely ...

# Pull request → merge: approved dish joins the official menu
git checkout main
git merge feature/new-risotto

# Revert: undo a bad change
git revert HEAD    # "go back to before I ruined the sauce"

# Tag: freeze a menu version
git tag v2.0 -m "Spring 2026 menu"
```

## 17. VS Code + Python + Virtual Environment

How the three layers work together:

```bash
# Layer 1: VS Code (the restaurant floor)
# → Where you write, run, and debug code
# → Has no cooking ability on its own

# Layer 2: Python installation (the building infrastructure)
# → Installed once on your machine
python --version       # Check: are the gas lines connected?

# Layer 3: Virtual environment (the isolated prep station)
python -m venv .venv           # Create a sealed-off station
source .venv/bin/activate      # macOS/Linux
.venv\Scripts\activate         # Windows

# Now pip installs go into THIS project's station only
pip install flask pandas       # project_a's ingredients
pip list                       # see what's in this station

# In VS Code: Ctrl+Shift+P → "Python: Select Interpreter"
# Point it to .venv/bin/python → everything is scoped
```

| Layer | Kitchen Analogy | What it provides |
|---|---|---|
| VS Code | Restaurant floor | Editor, terminal, debugger, extensions |
| Python | Building infrastructure | Interpreter, standard library, pip |
| Virtual env | Isolated prep station | Per-project packages, no conflicts |

````python
# DICOM CT example
# Install `pydicom` first if needed: pip install pydicom
import pydicom
import matplotlib.pyplot as plt
from pathlib import Path

dicom_path = Path("path/to/ct_slice.dcm")

ds = pydicom.dcmread(dicom_path)
print("Patient:", ds.get("PatientName", "Unknown"))
print("Study date:", ds.get("StudyDate", "Unknown"))
print("Modality:", ds.get("Modality", "Unknown"))
print("Image shape:", ds.pixel_array.shape)

intercept = float(getattr(ds, "RescaleIntercept", 0.0))
slope = float(getattr(ds, "RescaleSlope", 1.0))
hu_image = ds.pixel_array.astype(float) * slope + intercept

plt.figure(figsize=(6, 6))
plt.imshow(hu_image, cmap="gray", vmin=-1000, vmax=400)
plt.title(f"CT slice: {dicom_path.name}")
plt.axis("off")
plt.show()
````
